A database is a collection of **tables**. Each table is made up of **rows** of data, and each row has values for the same set of **columns**.

```python
import sqlite3
```

```python
conn = sqlite3.connect("jobs.db")
cursor = conn.cursor()
```

```python
# Use the query that returns the Rank and Major columns from the table recent_grads.
query = "SELECT Rank,Major FROM recent_grads;"
cursor.execute(query)
results = cursor.fetchall()

results
```

```python
query = "SELECT Rank,Major_code,Major,Major_category,Total FROM recent_grads;"
cursor.execute(query)
results = cursor.fetchall()

results
```

```python
# Run the query that returns the Major and ShareWomen values for all rows where ShareWomen exceeded 0.5.
query = "SELECT Major,ShareWomen FROM recent_grads WHERE ShareWomen>0.5;"
cursor.execute(query)
results = cursor.fetchall()

results
```

```python
# A SQL query that returns all majors with more than 10000 employed graduates.
query = "SELECT Major,Employed FROM recent_grads WHERE Employed>10000;"
cursor.execute(query)
results = cursor.fetchall()

results
```

```python
# A query that returns only the first 10 results.
query = "SELECT Major FROM recent_grads WHERE Employed>10000 LIMIT 10;"
cursor.execute(query)
results = cursor.fetchall()

results
```
