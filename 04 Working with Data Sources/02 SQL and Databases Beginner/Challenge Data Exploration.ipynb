```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("factbook.db")
```

```python
pd.read_sql_query("SELECT * FROM facts LIMIT 10;", conn)
```

See the [documentation on SQLite aggregation functions](https://www.sqlite.org/lang_aggfunc.html) if you get stuck.

```python
# In SQL, calculate the means of the population, population_growth, birth_rate, and death_rate columns.
result_turple = conn.execute("SELECT AVG(population),AVG(population_growth),AVG(birth_rate),AVG(death_rate) \
FROM facts;").fetchall()[0]

print(result_turple)

pop_avg = result_turple[0]
pop_growth_avg = result_turple[1]
birth_rate_avg = result_turple[2]
death_rate_avg = result_turple[3]
```

```python
# Calculate the minimum and maximum values for the columns from the previous cell.
result_turple = conn.execute("SELECT MIN(population),MAX(population),\
MIN(population_growth),MAX(population_growth),\
MIN(birth_rate),MAX(birth_rate),\
MIN(death_rate),MAX(death_rate) \
FROM facts;").fetchall()[0]

print(result_turple)

pop_min = result_turple[0]
pop_max = result_turple[1]
pop_growth_min = result_turple[2]
pop_growth_max = result_turple[3]
birth_rate_min = result_turple[4]
birth_rate_max = result_turple[5]
death_rate_min = result_turple[6]
death_rate_max = result_turple[7]
```

```python
# The max value for the population column is over 7 billion, however.
# The minimum value for the population column is also problematic, because no country has 0 people.
# These quirks exist because the database contains rows for entities that aren't countries. 
# There's a row representing the entire world, for example (hence the 7 billion population),
# and some rows representing oceanic areas (hence the population of 0).
pd.read_sql_query("SELECT * FROM facts WHERE population<100 OR population>1000000000;", conn)
```

```python
# A single query that returns the following minimum and maximum values for countries 
# where population is less than 2 billion and population is greater than 0.
result_turple = conn.execute("SELECT MIN(population),MAX(population),\
MIN(population_growth),MAX(population_growth),\
MIN(birth_rate),MAX(birth_rate),\
MIN(death_rate),MAX(death_rate) \
FROM facts \
WHERE population>0 AND population<2000000000;").fetchall()[0]

print(result_turple)

pop_min = result_turple[0]
pop_max = result_turple[1]
pop_growth_min = result_turple[2]
pop_growth_max = result_turple[3]
birth_rate_min = result_turple[4]
birth_rate_max = result_turple[5]
death_rate_min = result_turple[6]
death_rate_max = result_turple[7]
```

Now let's predict next year's population for each country using the following formula:  
**projected_population = population + (population * (population_growth/100))**

```python
# Use SQL arithmetic to return the projected population values using the above formula and the following parameters.
query = "SELECT name,population,population_growth,ROUND(population+(population*(population_growth/100))) \
AS projected_population \
FROM facts \
WHERE population IS NOT NULL AND population_growth IS NOT NULL \
AND population>0 AND population<2000000000;"
pd.read_sql_query(query, conn)
```

```python
# To understand how global population would shift under the projections,
# calculate the minimum, maximum, and average values.
query = "SELECT ROUND(MIN(population+(population*(population_growth/100)))) AS pop_proj_min,\
ROUND(MAX(population+(population*(population_growth/100)))) AS pop_proj_max,\
ROUND(AVG(population+(population*(population_growth/100)))) AS pop_proj_avg \
FROM facts \
WHERE population IS NOT NULL AND population_growth IS NOT NULL \
AND population>0 AND population<2000000000;"

result_turple = conn.execute(query).fetchall()[0]

print(result_turple)

pop_proj_min = result_turple[0]
pop_proj_max = result_turple[1]
pop_proj_avg = result_turple[2]

pd.read_sql_query(query, conn)
```
