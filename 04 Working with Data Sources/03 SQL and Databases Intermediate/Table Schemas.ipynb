```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("factbook.db")

pd.read_sql_query("SELECT * FROM facts LIMIT 10;", conn)
```

```python
query = 'PRAGMA table_info(facts);'
pd.read_sql_query(query, conn)
```

This definition, which contains information like column names, data types, and which column is the primary key, is called a **table schema**.

```python
# The code will add a column called awesomeness to the facts table.
conn.execute('ALTER TABLE facts ADD awesomeness integer;')

query = 'PRAGMA table_info(facts);'
pd.read_sql_query(query, conn)
```

When you add a column, all the values associated with it will be **NULL** to begin with.

```python
pd.read_sql_query("SELECT * FROM facts LIMIT 10;", conn)
```

Here is [the full list of the data type](https://www.sqlite.org/datatype3.html).

```python
conn.execute('ALTER TABLE facts ADD leader text;')

query = 'PRAGMA table_info(facts);'
pd.read_sql_query(query, conn)
```

The query would remove the column awesomeness from the table facts.  
**'ALTER TABLE facts DROP COLUMN awesomeness;'**   

This command is only possible in certain SQL versions, and isn't possible with the SQLite database engine, so we won't practice it right now.
